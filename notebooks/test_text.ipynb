```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(r'D:\Master_SAR\MLA\PJT\Git\Matcha-TTS-etu-UPMC-ENSAM')))
import re

from matcha.text_to_ID.cleaners import english_cleaners
from matcha.text_to_ID.symbols import symbols
from matcha.text_to_ID.fonctions import text_to_sequence
from matcha.text_to_ID.cmudict import CMUDict
```

```python
def tester_mon_texte(texte_brut):
    """
    Applique le pipeline de nettoyage complet.
    """
    texte_corrige = english_cleaners(texte_brut)
    
    print("-" * 30)
    print(f"TEXTE D'ENTRÉE : {texte_brut}")
    print(f"TEXTE CORRIGÉ : {texte_corrige}")
    print("-" * 30)
    
    if any(char.isdigit() for char in texte_brut) and not any(char.isdigit() for char in texte_corrige):
        print("Info : Les nombres ont été correctement convertis en lettres.")
    
    return texte_corrige
```

```python
mon_texte = "Hello! I have 2 apples and Mr. Smith has 10.5 dollars."

resultat = tester_mon_texte(mon_texte)
```

```python
texte = "Hello world, I have 3 apples."
sequence = text_to_sequence(texte)

CMUDICT_PATH = "D:\Master_SAR\MLA\PJT\Git\Matcha-TTS-etu-UPMC-ENSAM\matcha\\text_to_ID\cmudict-0.7b"
cmu_dict = CMUDict(CMUDICT_PATH)
# Regex pour extraire les mots (avec apostrophes) et la ponctuation
_token_re = re.compile(r"([a-zA-Z']+)|([.,!?;:])")
cleaned_text = english_cleaners(texte)
sequence_phoneme = []
for token in _token_re.findall(cleaned_text):
    word, punct = token  # token est un tuple (mot, ponctuation)
    
    if word:  # Si c'est un mot
        pronunciation = cmu_dict.lookup(word.upper())
        sequence_phoneme.append(pronunciation)
print(texte,sequence_phoneme, sequence)
```
